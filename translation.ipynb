```python
from torch.nn.modules import padding
#create language translation using hugging face from transformers
from transformers import MarianMTModel, MarianTokenizer
#To select the model name from hugging face transformers
model_name = 'Helsinki-NLP/opus-mt-en-hi'
#To load the model using model name
model = MarianMTModel.from_pretrained(model_name)
#To load the tokenizer using model name
tokenizer = MarianTokenizer.from_pretrained(model_name)
#To e input text from user to translate
prompt = input('Enter text to translate: ')
#To encode the input text into input tokens using tokenizer
input_tokens = tokenizer(prompt, return_tensors='pt', padding = True)
#To generate the output tokens using model and input tokens
output_tokens = model.generate(**input_tokens)
#To decode the output tokens into translated text
print(tokenizer.decode(output_tokens[0], skip_special_tokens=True))
#Upload the code to github
```
